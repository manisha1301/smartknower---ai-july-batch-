### 2) Resizing

To resize, we need different interpolation methods
1) cv2.INTER_AREA - Used to shrink an image<br>
2) cv2.INTER_LINEAR - Used to expand an image<br>
2) cv2.INTER_CUBIC - Can be used for both<br>

```python
b = cv2.imread('apple.jpg')
b = cv2.cvtColor(b,cv2.COLOR_BGR2RGB)
plt.imshow(b)
plt.show()
```

```python
b.shape
```

```python
r1 = cv2.resize(b,(300,300),cv2.INTER_LINEAR)
r2 = cv2.resize(b,(70,70),cv2.INTER_AREA)
r3 = cv2.resize(b,(175,350),cv2.INTER_CUBIC)
r4 = cv2.resize(b,(350,175),cv2.INTER_CUBIC)

plt.imshow(b),plt.title('Original'),plt.show()
plt.imshow(r1),plt.title('Expanded,zoomed in'),plt.show()
plt.imshow(r2),plt.title('Shrinked'),plt.show()
plt.imshow(r3),plt.title('both_1'),plt.show()
plt.imshow(r4),plt.title('both_2'),plt.show()

```

### Thresholding

Thresholidng is connverting greyscale images into its binary format

### Methods of Thresholding

Assumed threshold is t

1) <b>cv2.THRESH_BINARY</b><br>
Converts pixels below t to 0(black) and above t to 255(white).

2) <b>cv2.THRESH_BINARY_INV</b><br>
Converts pixels below t to 255(white) and above t to 0(black)

3) <b>cv2.THRESH_TRUNC</b><br>
Pixels below t remain unchanged and pixels aboe t are truncated (held at t).

4) <b>cv2.THRESH_TOZERO</b><br>  Pixels below t are converted to 0(black) and pixels above t held as it is (remain unchanged)

5) <b>cv2.THRESH_TOZERO_INV</b><br>  Pixels above t are converted to 0(black) and pixels nbelow t held as it is (remain unchanged)

```python
m = cv2.imread('kneau.jpg',0)
plt.imshow(m,'gray')
plt.show()
```

```python
m.shape
```

```python
m[30]
```

```python
# Converts pixels below t to 0(black) and above t to 255(white).
t = 110
_,t1 = cv2.threshold(m,t,255,cv2.THRESH_BINARY)
plt.imshow(t1,'gray')
plt.show()
```

```python
t1[30]
```

```python
# Converts pixels below t to 255(white) and above t to 0(black).
t = 110
_,t2 = cv2.threshold(m,t,255,cv2.THRESH_BINARY_INV)
plt.imshow(t2,'gray')
plt.show()
```

```python
t2[30]
```

```python
# Pixels below t remain unchanged and pixels aboe t are truncated (held at t).
t = 110
_,t3 = cv2.threshold(m,t,255,cv2.THRESH_TRUNC)
plt.imshow(t3,'gray')
plt.show()
```

```python
t3[30]
```

```python
# Pixels below t are converted to 0(black) and pixels above t held as it is (remain unchanged)
t = 110
_,t4 = cv2.threshold(m,t,255,cv2.THRESH_TOZERO)
plt.imshow(t4,'gray')
plt.show()
```

```python
t4[30]
```

```python
# Pixels above t are converted to 0(black) and pixels below t held as it is (remain unchanged)
t = 110
_,t5 = cv2.threshold(m,t,255,cv2.THRESH_TOZERO_INV)
plt.imshow(t5,'gray')
plt.show()
```

```python
t5[30]
```

### Image Segmentation based on Colors

```python
s1 = cv2.imread('s1.jpg')
s1 = cv2.cvtColor(s1,cv2.COLOR_BGR2RGB)
plt.imshow(s1)
plt.show()
```

```python
def rgb2hex(rgb):
    hex_img = '#{:02x}{:02x}{:02x}'.format(int(rgb[0]),int(rgb[1]),int(rgb[2]))
    return hex_img
```

```python
rgb2hex([255,0,0])
```

```python
rgb2hex([0,255,0])
```

```python
rgb2hex([0,0,255])
```

```python
from sklearn.cluster import KMeans
```

```python
s1.shape
```

```python
s1_re = s1.reshape(s1.shape[0]*s1.shape[1],s1.shape[2])
s1_re.shape
```

```python
k4 = KMeans(n_clusters=4)
k4.fit(s1_re)

ypred4 = k4.predict(s1_re)
print(ypred4)
```

```python
print(len(ypred4))
```

```python
from collections import Counter
c1 = Counter(ypred4)
print(c1)
```

```python
cent4 = k4.cluster_centers_
print(cent4)
```

```python
col_in_img = [rgb2hex(cent4[i]) for i in range(4)]  # i=0,1,2,3
col_in_img
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(10,10))

ax1.imshow(s1)
ax1.set_title('Original Image')

ax2.pie(c1.values(),labels=col_in_img,colors=col_in_img,autopct="%.3f%%")
ax2.set_title('Segmenting image into top 4 colors')

plt.show()
```
