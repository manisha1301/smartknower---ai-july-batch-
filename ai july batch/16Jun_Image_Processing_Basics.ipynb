2) Grayscale Images<br>

### Image Inpainting

```python
a3 = cv2.imread('kneau.jpg')
plt.imshow(a3)
plt.show()
```

```python
s = cv2.rectangle(a3,(10,20),(50,40),(255,0,0),-1)  # filled rectangle
s = cv2.circle(a3,(20,75),15,(0,255,0),2)           # hollow circle with border thickness
s = cv2.putText(a3,'Kneau',(5,125),cv2.FONT_HERSHEY_SIMPLEX,2,(0,0,255),2)
plt.imshow(s)
plt.show()
```

### Creating an image using numpy Array

```python
d = np.zeros((300,300,3),dtype=np.uint8)
d[10:40,20:50] = [255,0,0]   # RGB  - Red
d[50:80,20:60] = [0,255,0]   # RGB  - Green
d[10:40,70:100] = [0,0,255]   # RGB  - Blue
d[50:80,70:100] = [255,255,0]   # RGB  - Yellow
d[100:150,100:180] = [255,255,255]   # RGB  - white
plt.imshow(d)
plt.show()
```

#### Blurring

```python
g = cv2.imread('flower.jpg')
g = cv2.cvtColor(g,cv2.COLOR_BGR2RGB)
plt.imshow(g)
plt.show()
```

```python
k1 = np.ones((3,3))/9
k2 = np.ones((5,5))/25
k3 = np.ones((7,7))/49

res1 = cv2.filter2D(g,-1,k1)
res2 = cv2.filter2D(g,-1,k2)
res3 = cv2.filter2D(g,-1,k3)

plt.imshow(g),plt.title('Original Image');plt.show()
plt.imshow(res1),plt.title('Blurred Image - 3x3 kernel');plt.show()
plt.imshow(res2),plt.title('Blurred Image - 5x5 kernel');plt.show()
plt.imshow(res3),plt.title('Blurred Image - 7x7 kernel');plt.show()
plt.show()
```

#### Sharpening and Embossing

```python
ker_sharp = np.array([[-1,-1,-1],[-1,9,1],[-1,-1,-1]])  # sharpening
ker_emboss = np.array([[0,-1,-1],[1,0,-1],[1,1,1]])     # emboss
r1 = cv2.filter2D(g,-1,ker_sharp)
r2 = cv2.filter2D(g,-1,ker_emboss)

plt.imshow(g),plt.title('Original Image');plt.show()
plt.imshow(r1),plt.title('Sharpened Image');plt.show()
plt.imshow(r2),plt.title('Embossed Image');plt.show()
plt.show()
```
